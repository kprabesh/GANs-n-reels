<h5> Motivation: </h5>

Our research aims to discover if successful AI image generative models can also be used to generate music. We plan to use a Generative Adversarial Network, a particular type of Artificial Neural Network, in an attempt to generate Irish music. 

Irish music, as one of the primary influences on traditional Newfoundland music, tends toward straightforward melodies with a fixed format. It is a style intended to be played with a single instrument by amateur musicians, resulting in notation that is simple and easy to follow. These unique qualities are the primary motivation for choosing Irish music as our target.

<h5> Background and Related Work: </h5> 

An Artificial Neural Network, or ANN, is based on a collection of connected units called nodes. These nodes can be loosely compared to neurons in the human brain with respect to their interconnectedness and influence on their neighbors. There are various types of ANNs, many of which have been successfully used to classify, predict, and generate different types of data. Recently, a UK-based team has used a Recurrent Neural Network (RNN) to generate music, and another team from Google Brain has generated music using a Long-Short Term Memory RNN. An important fact to consider is that these two models do not generate music from random noise; they generate music from a small sequence of notes, and ‘predict’ what should come next.  


A Generative Adversarial Network, or a GAN, is an ANN consisting of a Generator that generates data when fed noise, and a Discriminator which classifies data. A common analogy is the Generator is akin to an art forger, while the Discriminator is like a detective. The analogy is useful as, after being trained, the Generator can be used to create original ‘art’. GANs have worked well with pictures, but to our knowledge there has not been a successful attempt to generate music using GANs.  

<h5> Research Approach: </h5>

We hope to use the distinctive structure of Irish tunes to make them suitable for GAN-based music generation. Our main idea is to regard an Irish melody as a fixed-size object with cross-references among its parts: “music as a picture” view. The two key components of this project are:

<ol>
  <li> Preprocessing: <ul>
      <li>Creating a well-defined format for music encoding suitable for GANs</li>
      <li>Representing “vertical” dependencies in the tunes</li> </ul> </li>
  <li> Modifying GANs for improved performance via:<ul>
      <li>Statistical “Bagging” and “Boosting” to mimic the full input distribution </li>
      <li>Modifying training rate to avoid “Mode Collapse”</li> </ul> </li> </ol>

<h5> Novelty: </h5>

Though there are RNN-based Irish music generators, we found no GANs which successfully generated music. Unlike RNNs, GANs could truly generate music from scratch, rather than using prediction. To our knowledge, no one has presented the idea of organizing 
music like a picture, which makes our encoding unique.

### Irish Music ###

Irish traditional music is the indigenous music of the Ireland. In spite of emigration and a well-developed connection to music influences from Britain and the United States, Irish traditional music has kept many of its elements and itself influenced many forms of music, such as the folk music in Newfoundland and Labrador. 

Irish traditional music includes many kinds of vibrant songs, including drinking songs, ballads and laments, sung unaccompanied or with accompaniment by a variety of instruments. Traditional dance music includes reels, hornpipes and jigs. In the nineteenth century folk instruments would have included the flute, the fiddle and the uilleann pipes.

We chose Irish music because:
1. It has a simple and distinctive structure.
2. We have access to a huge database of over 30,000 Irish songs through The Sessions website.
3. Irish songs can be represented using ABC notation, which is vectorized easily.

### Viewing Music as a Picture ###

Our main idea is to regard an Irish melody as a fixed-size object with cross-references among its parts. We call this a “music as a picture” view. To do this, we had to create a well-defined format for music encoding suitable for GANs, and define the process to create this format.

The irish songs in our dataset are in abc notation: which is a popular and simple way to represent tunes in plain text. To prepare this data for a Generative Adversarial Network, we transformed this notation to a vector, using the viewing music as a picture process. 

This process is divided into 4 stages defined as: 
1. Conversion: converting abc notation strings into a machine readable format. 
2. Filtering: removing entries from the dataset entirely that are deemed unsuitable as training data
3. Cleaning: removing or modifying elements of the abc string itself in an even simpler format
4. Vectorizing: converting cleaned strings into matricies of numbers

### Common Musical Notation Symbols ###

Musical notation and symbols have been in use since the 13th century. They describe the pitch, rhythm, tempo, and articulation of music. Some of the common musical symbols encountered in ABC notation of Irish music include:

* Musical note: which represents the pitch and duration of a sound
* Key signatures: define the prevailing key of the music that follows. The most common signatures for Irish music are G major and D major
* Time signatures: establish the number of beats in each bar. Most common ones for Irish music are 4/4 time signature.
* Bar lines: which indicate changes in time signatures
* Braces: which connect two or more music notes that are played simultaneously
* Accidentals: which modify the pitch of the notes that follow them. common ones are:
 * flat: lowers the pitch of a note by one semitone
 * sharp: raises the pitch of a note by one semitone
 * natural: cancels a previous accidental
* Note relationships: indicate the relation between two or more notes joined together. Common ones include:
 * tie: identical notes that are played as a single note
 * slur: two different notes played in one smooth motion
* Ornaments: which modify the pitch pattern of individual notes. common ones include:
 * trill: a rapid alteration between the specified note and the next higher note
* Repeat signs: enclose a passage that is to be played more than once

### ABC Notation ###

ABC is a great way of passing on tunes by writing the notes out in plain text. It's very straightforward and easy to learn. More information can be found [here](http://abcnotation.com/blog/2010/01/31/how-to-understand-abc-the-basics/)

Typically a tune consists of two parts: the **header** which contains background information, and the **notes** which contain the notes.

The **header** consists of different fields which describe
* title: the name of the song
* composer: the name of the composer of the song
* meter: refers to the time signature of the song 
* key: refers to the default group of notes that can form the song

The **notes** of the tune consist of A, B, C, D, E, F, G which represent the musical note. Upper case letters are used to denote the bottom octave, while lower case represent the top octave. An apostrophe references going up an octave while a comma denotes going down an octave.

The **note length** is determined by a separate field, and a number after each note determines how long a note should be held for. A /x refers to shortening a note by the appropriate number x. If no note length is given, the default note length is determined in the header.

**Bar lines** are easily inserted with a | symbol and you can use || for a double bar line or |] for a thin-thick double bar line. Similarly |: represents the start of a repeated section and :| represents the end. Where one repeated section ends and a new one starts, use :||: or :: for short.


### Creating a Context-Free Grammar from ABC Notation ###


While ABC notation is easy to use, it is inherently a context-sensitive grammar. This is apparent since in the D major key, F represents F# implicitly, while in the C major key it represents F. There are also different ways to encode the same information, for example, a trill can be defined using ! or + or T, which mean the same thing in ABC notation. The subdivision of beats is also not explicit. In order to feed this data so a Neural Network can be trained on, we need to simplify this data much further and make it more objective. 

To convert this into a context-free grammar, we have done the following:
* kept the relative distance between nodes, by transposing the key signature to D, and mapped the resulting ABC notation to a number
* kept the temporal information explicit, by ensuring each bar to have a constant number of beats and extending the bar size 
* created a separate time vector to explicitly differentiate between pressing a note vs holding it.
* removed all ornamentation and other musical symbols, explicitly stating them when needed, such as adding notes for repeats.

Doing all this, we can create a context-free grammar out of abc notation, which makes training on a GAN easier.

### Conversion ###

### Filtering ### 

To create a fixed-size vector, we used only a 4/4 time signature to ensure all bars have the same note length. Discarding all other time stamps, we were left with 16,000 songs out of 30,000.

Since the Sessions allows for anyone to upload music, we found some human errors in the encoding of music. Any song that was inconsistent musically (eg. has less than 16 bars, has unknown ornaments, etc) was filtered out.

We were left with around 10,000 songs in standard abc notation after our filtering process.

### Cleaning ###

As part of the cleaning process, we converted the tunes with standard abc notation to our context-free abc grammar.

Some of the examples in cleaning include:

```python
def remove_quotations(frame):
    """
    Chord symbols are sypically placed within quotation marks
    """
    frame['abc'] = frame['abc'].str.replace('".*?"', '')
    return frame
```

```python
def consolidate_barlines(frame):
    """
    Turns all barlines and repeatlines into single '|' characters
    see http://abcnotation.com/wiki/abc:standard:v2.1#repeat_bar_symbols
    """
    frame['abc'] = frame['abc'].str.replace('\|+|\|:|\|[|:\||]\|', '|')
    return frame
```

```python
class GAN():
    def __init__(self):
        self.img_dim = [4+1, 64+(4*2)] #RHS of sum is padding
        self.channels = 1
        self.img_shape = [*self.img_dim, self.channels]
        self.noise_shape = [100, ]

        self.gloss = []
        self.dloss = []
        self.intervals = []
        self.graphInterval = 50

        optimizer = optimizers.Adam(0.0002, 0.5)

        self.discriminator = self.build_discriminator()
        self.discriminator.compile(loss='binary_crossentropy',
                                   optimizer=optimizer,
                                   metrics=['accuracy'])

        self.generator = self.build_generator()

        noise = layers.Input(shape=self.noise_shape)
        img = self.generator(noise)

        self.discriminator.trainable = False

        valid = self.discriminator(img)

        self.combined = models.Model(inputs=noise, outputs=valid)
        self.combined.compile(loss='binary_crossentropy', optimizer=optimizer)

    def build_generator(self, momentum=0.8, alpha_leak=0.2):
        noise_shape = self.noise_shape

        model = models.Sequential()

        model.add(layers.Dense(1024, input_shape=noise_shape))
        model.add(LeakyReLU(alpha=alpha_leak))
        model.add(layers.BatchNormalization(momentum=momentum))
        
        model.add(layers.Dense(64*(1+self.img_dim[0]//2)*(self.img_dim[1]//2))) #64 filters, each of half the size of the input image
        model.add(LeakyReLU(alpha=alpha_leak))
        model.add(layers.BatchNormalization(momentum=momentum))
        model.add(layers.Reshape([(1+self.img_dim[0]//2),(self.img_dim[1]//2), 64]))
        
        model.add(layers.Conv2DTranspose(32, [2,5], padding="same")) #32 filters, each of the size of the input image
        model.add(LeakyReLU(alpha=alpha_leak))
        model.add(layers.BatchNormalization(momentum=momentum))
        model.add(layers.UpSampling2D(size=[2,2])) #now the filters are of the full size of the image, with 1 additional row at the top
        model.add(layers.Cropping2D(cropping=[[1, 0], [0, 0]])) #remove extra padding at the top
        
        model.add(layers.Conv2DTranspose(1, [2,5], padding="same", activation="tanh")) #image

        model.summary()

        noise = layers.Input(shape=noise_shape)
        img = model(noise)

        return models.Model(noise, img)

    def build_discriminator(self, alpha_leak=0.2):
        img_shape = self.img_shape

        model = models.Sequential()

        model.add(layers.Conv2D(32, [2,9], strides=[1,1], padding="valid", input_shape=img_shape))
        model.add(layers.Cropping2D(cropping=[[0, 1], [4, 4]]))
        model.add(LeakyReLU(alpha=alpha_leak))
        model.add(layers.Conv2D(64, [2,9], strides=[2,2], padding="same"))
        model.add(LeakyReLU(alpha=alpha_leak))
        model.add(layers.Flatten())
        model.add(layers.Dense(1024))
        model.add(LeakyReLU(alpha=alpha_leak))
        model.add(layers.Dense(1, activation="sigmoid"))

        model.summary()

        img = layers.Input(shape=img_shape)
        validity = model(img)

        return models.Model(img, validity)

    def train(self, iterations, batch_size=60, sample_interval=750):
        X_train = paddedData
        halfMaxPitch = (80+53)//2
        pitchRange = 80 - halfMaxPitch
        X_train = (X_train.astype(np.float32) - halfMaxPitch) / pitchRange
        X_train = np.expand_dims(X_train, axis=3)

        for iteration in range(iterations+1):
            for _ in range(1): #train discriminator more times
              discriminator_loss = self.train_discriminator(X_train, batch_size=batch_size)
              
            generator_loss = self.train_generator(batch_size)

            if(iteration % self.graphInterval):
                self.dloss.append(discriminator_loss[0])
                self.gloss.append(generator_loss)
                self.intervals.append(iteration)

            if iteration % sample_interval == 0:
                print("{} [D loss: {}, acc.: {:.2f}%] [G loss: {}]".format(
                    iteration,
                    discriminator_loss[0],
                    100 * discriminator_loss[1],
                    generator_loss))
                self.sample_images()

        #self.plotLossHistory()

    def train_discriminator(self, X_real, batch_size):
        half_batch = batch_size // 2

        discriminator_indices = np.random.randint(0, X_real.shape[0], half_batch)
        discriminator_train_imgs = X_real[discriminator_indices]

        noise = np.random.normal(0, 1, [half_batch, 100])
        generated_imgs = self.generator.predict(noise)

        discriminator_loss_real = self.discriminator.train_on_batch(discriminator_train_imgs, np.ones([half_batch, 1]))
        discriminator_loss_gen = self.discriminator.train_on_batch(generated_imgs, np.zeros([half_batch, 1]))
        # average the two losses
        discriminator_loss = np.add(discriminator_loss_real, discriminator_loss_gen) / 2

        return discriminator_loss

    def train_generator(self, batch_size):
        noise = np.random.normal(0, 1, [batch_size, 100])
        valid_y = np.array([1] * batch_size)  # move generator towards desired validity

        generator_loss = self.combined.train_on_batch(noise, valid_y)
        return generator_loss
      
    def plotLossHistory(self):
        plt.plot(self.intervals, self.dloss)
        plt.plot(self.intervals, self.gloss)
        plt.legend(["Discriminator Loss", "Generator Loss"])
        plt.xlabel("Iteration")
        plt.ylabel("Loss")
        plt.ylim(ymin=0, ymax=5)
        plt.show()

    def sample_images(self):
        rows, columns = 4, 4
        noise = np.random.normal(0, 1, [rows * columns, 100])
        
        generated_imgs = self.generator.predict(noise)[:, :-1, 4:-4]
        generated_imgs = 0.5 * generated_imgs + 0.5
        
        random_real_indices = np.random.randint(0, data.shape[0], rows*columns)
        real_imgs = data[random_real_indices]
                                        
        print("Real")
        fig, axs = plt.subplots(rows, columns)
        for row in range(rows):
            for column in range(columns):
                count = row * columns + column
                axs[row, column].imshow(real_imgs[count, :, :], cmap='gray', aspect='auto')
                axs[row, column].axis('off')

        plt.show()
                                        
        print("Generated")
        fig, axs = plt.subplots(rows, columns)
        for row in range(rows):
            for column in range(columns):
                count = row * columns + column
                axs[row, column].imshow(generated_imgs[count, :, :, 0], cmap='gray', aspect='auto')
                axs[row, column].axis('off')

        plt.show()
        self.plotLossHistory()


gan = GAN()
gan.train(iterations=4000, sample_interval=200)
```

```python
noise = np.random.normal(0, 1, [20, 100]) #20 arrays of noise of shape [100,]
generated_samples = gan.generator.predict(noise)
```

```python
generated_samples = generated_samples[:,:-1, 4:-4]
generated_samples = np.squeeze(generated_samples, axis=3) #Remove colour channel
generated_samples = generated_samples.reshape([-1, 16, 16])
```

```python
MAX_PITCH = 80
MIN_PITCH = 53
MID_PITCH = (MAX_PITCH + MIN_PITCH) / 2
RANGE = MAX_PITCH - MIN_PITCH

#Map from [-1, 1] to [MIN_PITCH, MAX_PITCH]
generated_samples = (RANGE * generated_samples) + MID_PITCH

#Make values discrete
generated_samples = generated_samples.astype(np.int32)

print("Shape: {}".format(generated_samples.shape))
print("First 2 scaled samples:")
print(generated_samples[:2])
```

```python
notes = {60: 'C', 61: 'D', 62: 'E', 63: 'F', 64: 'G', 65: 'A', 66: 'B', 67: 'c', 68: 'd', 69: 'e', 70: 'f', 71: 'g', 72: 'a', 73: 'b'}

def nums_to_abc(nums):
    chars = []
    for num in nums:
        if num in notes:
            chars.append(notes[num])
        elif num > 73:
            apostrophe_count = np.ceil((num - 73) / 7).astype(np.int32)
            apostrophe_count = np.asscalar(apostrophe_count)
            chars.append(notes[num - 7*apostrophe_count])
            chars.extend(["'"] * apostrophe_count)
        elif num < 60:
            comma_count = np.ceil((60 - num) / 7).astype(np.int32)
            comma_count = np.asscalar(comma_count)
            chars.append(notes[num + 7*comma_count])
            chars.extend([","] * comma_count)
            
    return "".join(chars)
  
def tune_to_abc(tune):
  bars = [nums_to_abc(bar) for bar in tune]
  return "|".join(bars)

tunes = [tune_to_abc(tune) for tune in generated_samples]

print("First 3 tunes")
print(tunes[:3])
```
